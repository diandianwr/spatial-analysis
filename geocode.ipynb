Using 'request' to interact with API services 

```python
import requests
geocode_url = 'http://maps.googleapis.com/maps/api/geocode/json'
address = {'address': 'psrc,seattle'}
r = requests.get(geocode_url, params = address)
results = r.json()['results']
location = results[0]['geometry']['location']
location
```

Using 'geocoder' with multiple online providers such as Google & Bing, some of them requre key, others have daily request limit. They also have different JSON responses between each other,
https://pypi.python.org/pypi/geocoder

```python
# pip install geocoder
import geocoder 
r = geocoder.google('psrc,seattle')
#r.json
print (r.json['lat'],r.json['lng'],r.json['address'])  
```

Working with address datasets
read the address into address_list and put all the locations into location_list

```python
location_list = []
for address in address_list:
    address = address + ", Seattle, WA, USA" #optional
    r = geocoder.google(address)
    location_list.append([r.json['lat'],r.json['lng']],r.json['address'])
```

Write geocoding result into address datasets

```python
import csv
with open('addresses.csv', 'wb') as csvfile:
    writer = csv.writer(csvfile, delimiter=',',dialect='excel')
    writer.writerow(['address','lat','lng'])
    for i in range(len(address_list)):
        writer.writerow([address_list[i],location_list[i][0],location_list[i][1]])
```
